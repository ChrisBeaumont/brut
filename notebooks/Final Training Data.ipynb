# Final Training Data

```python
%pylab

import json
import random
import itertools
import os
import sys

import numpy as np
from sklearn.metrics import auc_score
from sklearn.utils import resample

from bubbly.wiserf import WiseRF
from bubbly.extractors import MultiViewExtractor, ManyManyExtractors
from bubbly.util import roc_curve
from bubbly.dr1 import WideLocationGenerator, highest_quality_on_params
from bubbly.model import Model
from bubbly.bootstrap import bootstrap_negatives

from bubbly.hyperopt import fmin, rf_space, auc_below_fpos
```

```python
params = {'max_features': 'auto', 'min_samples_split': 4, 'n_jobs': 2, 'criterion': 'infogain', 'n_estimators': 800}

def score(onval, offval):
    y = [1] * onval.shape[0] + [0] * offval.shape[0]
    yp = np.hstack([onval, offval])
    return auc_below_fpos(y, yp, .002)

def build_model(mod3):
    l = WideLocationGenerator(mod3)
    ex = MultiViewExtractor(ManyManyExtractors())
    ex.shp = (60, 60)
    clf = WiseRF(**params)
    return Model(ex, l, clf)

def locations(model):
    
    pos = filter(lambda x: model.locator.valid_longitude(x[0]), highest_quality_on_params())
    cv_pos = filter(lambda x: not model.locator.valid_longitude(x[0]), highest_quality_on_params())
                 
    neg = list(itertools.islice(model.locator.negatives_iterator(samples_per_field=500), 0, 15000))
    
    return sorted(pos), sorted(neg), sorted(cv_pos)

def initial_training_data(pos, neg):
    return sorted(pos), sorted(random.sample(neg, 3 * len(pos)))

def show_progress(dfs):
    pos, neg = dfs['cv_pos'], dfs['neg']
    roc_curve([0] * neg.size + [1] * pos.size, np.hstack((neg, pos)), lw=3)

def bootstrap_optimize(model, pos, neg, cv_pos, rounds=3):
    
    result, best_score = None, -np.inf
    
    tron, troff = initial_training_data(pos, neg)
    
    for i in range(rounds):
        print "Fit"
        
        sys.stdout.flush()
        model.fit(tron, troff)
        
        print "Decision Functions"
        sys.stdout.flush()
        
        dfs = {}
        df_func = lambda x: model.cloud_decision_function(x, workers=50 if len(x) > 500 else 5)
        dfs['cv_pos'] = df_func(cv_pos)
        dfs['neg_old'] = df_func(troff)
        dfs['neg'] = df_func(neg)
        
        show_progress(dfs)
        xlim(0, .002)
        show()
        
        s = score(dfs['cv_pos'], dfs['neg'])
        if s > best_score:
            result = {'pos' : tron, 'neg': troff}
            yield result
            best_score = s

        print "Round %i: %f" % (i, s)
        sys.stdout.flush()
        
        troff = bootstrap_negatives(dfs, neg, troff)


def main(i):
    model = build_model(i)
    pos, neg, cv_pos = locations(model)
    for result in bootstrap_optimize(model, pos, neg, cv_pos):
        outfile = '../models/training_data_%i.json' % i
        with open(outfile, 'w') as out:
            json.dump(result, out)
    return result
```
