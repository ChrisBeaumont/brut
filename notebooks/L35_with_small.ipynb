## Cross Matching MWP with Bubbly

```python
def xmatch(a, b):
    a = np.asarray(a)
    b = np.asarray(b)
    #distance matrix
    d = np.hypot(a[:, 1].reshape(-1, 1) - b[:, 1].reshape(1, -1),
                 a[:, 2].reshape(-1, 1) - b[:, 2].reshape(1, -1))
    ra = a[:, -1].reshape(-1, 1)
    rb = b[:, -1].reshape(1, -1)
    bad = (d > np.minimum(ra, rb))
    d[bad] = np.inf
    
    ratio = np.maximum(ra / rb, rb / ra)
    bad = ratio > 2
    d[bad] = np.inf
    
    result = np.zeros(a.shape[0], dtype=np.int) - 1
    
    while np.isfinite(d).any():
        best = np.argmin(d)        
        i, j = np.unravel_index(best, d.shape)
        
        result[i] = j
        d[i, :] = np.inf
        d[:, j] = np.inf
        
    return result
```

```python
match = xmatch(merged, mwp)
print 'Bubbly -> mwp: (%i)' % ((match >= 0).sum())
for i, j in enumerate(match):    
    if j < 0:
        continue
    print '%i -> %i' % (i, j)
    
print 'Unmatched Bubbly (%i / %i)' % ((match == -1).sum(), merged.shape[0])
print np.where(match == -1)[0].tolist()

print 'Unmatched mwp (%i / %i)' % (len(mwp) - (match >= 0).sum(), len(mwp))
z = np.zeros(len(mwp))
z[match] = 1
print np.where(z == 0)[0].tolist()
```

```python
import ds9
import os
from bubbly.field import Field

class BubblyViewer(object):
    def __init__(self):
        self.ds9 = None
        
    def start(self):
        if self.ds9 is None:
            self.ds9 = ds9.ds9()
        
    def load_longitude(self, lon):
        self.start()
        
        f = Field(lon)
        g = os.path.join(f.path, 'registered', '%3.3i_i4.fits' % lon)
        r = os.path.join(f.path, 'registered', '%3.3i_mips.fits' % lon)
        self.ds9.set('frame delete')
        self.ds9.set('frame new rgb')
        self.ds9.set('rgb red')
        self.ds9.set('file %s' % r)
        self.ds9.set('rgb green')
        self.ds9.set('file %s' % g)
        
        self._set_zscale()
        self._align_galactic()
        
    def _set_zscale(self):
        self.ds9.set('rgb red')
        self.ds9.set('scale asinh')
        self.ds9.set('rgb green')
        self.ds9.set('scale asinh')
        
    def _align_galactic(self):
        self.ds9.set('wcs galactic')
        self.ds9.set('wcs skyformat degrees')
        
    def look_at(self, params):
        if self.ds9 is None:
            self.load_longitude(params[0])
            
        l, b = params[1:3]
        r = params[-1]
        
        #this is a hacky guess
        zoom = 2 / 3600. / r * 500
        
        self.ds9.set('pan to {l} {b} galactic'.format(l=l, b=b))
        self.ds9.set('zoom to %f' % zoom)
        
    def outline(self, params, color='blue'):
        if not hasattr(params[0], '__len__'):
            params = [params]
            
        for i, p in enumerate(params):
            l, b, r = p[1:]
            self.ds9.set('regions', 'galactic; circle(%f,%f,%f)#color=%s text="%s"' % (l, b, r, color, i))            
            
    def clear(self):
        self.ds9.set('regions delete all')
```

## Summary, when using both small and large stamps


Bubbly finds twice as many sources as the MWP catalog (62 vs 32). 58% are real, 42% are not

* 27% of these (17 / 62) match unambiguously to a MWP source
* A futher 21% of these (13 / 62) "ambiguously" match to a MWP source -- that is, one or more MWP bubbles are split up into different regions.
* 10% are new bubbles not in the MWP project
* 26% are regions confused by extended / saturated 24 um emission
* 16% are regions of nebulosity, that don't look particularly shell-like

In addition, ignoring confusion examples, there are 12 bubbles in the MWP with no bubbly counterpart:

 * (6) are false positives -- mostly other nebulosity
 * (3) are true positives, wrongly rejected by bubbly
 * (3) are at the image edges, and un-classifiable by bubbly
 
 
Assuming this accounts for all detectable bubbles, then bubbly recovers 30 / 34 bubbles (88%), at a false positive rate of 42%.

The MWP, on the other hand, recovers 28 / 34 bubbles (82%), at a false positive rate of 19%.
